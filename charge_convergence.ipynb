```python
%load_ext autoreload
%autoreload 2

import matplotlib.pyplot as plt
import numpy as np

from gpaw import GPAW, PW
from ase import Atoms
from potentials import ParameterizedPotential
from gpaw.utilities.ps2ae import PS2AE
import utils
```

```python
Z = 1

n0 = []
for gpts in [60,70,80,90,100,110]:
    atoms = Atoms([Z], [(0, 0, 0)])
    atoms.center(vacuum=3)

    calc = GPAW(mode=PW(600), hund=True, eigensolver = 'cg', gpts=(gpts,)*3)
    atoms.set_calculator(calc)

    atoms.get_potential_energy()
    
    n = calc.get_all_electron_density(gridrefinement=1)
    
    n0.append(n.max())
```
